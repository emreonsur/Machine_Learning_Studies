```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/emreo/OneDrive/Desktop/Machine_Learning_Studies/Tutorials/Youtube_Tutorials/codebasics/Dummy_Variables_and_One_Hot_Encoding/Exercise/car_prices.csv")
df
```

```python
dummies = pd.get_dummies(df.get('Car Model'))
dummies
```

```python
merged = pd.concat([dummies, df], axis='columns')
merged
```

```python
final = merged.drop(['Car Model', 'Audi A5'], axis='columns')
final
```

```python
X = final.drop(['Sell Price($)'], axis='columns')
X
```

```python
y = final.get('Sell Price($)')
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
merged
```

# 1) Prediction of the price of a mercedez benz that is 4 yr old with mileage 45000

```python
model.predict([[0, 1, 45000, 4]])
```

# 2) Prediction of the price of a BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[1, 0, 86000, 7]])
```

# 3) score (accuracy) of the model

```python
model.score(X, y)
```
